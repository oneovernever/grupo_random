1. Dado el registro de notas de los alumnos de la forma (padr√≥n, materia, nota, fecha). Se pide resolver utilizando Pandas:

```
df = pd.DataFrame(columns=['padron','materia','nota','fecha'])
```

```
df.head()
```

```
# Generemos notas aleatoreas.
df.nota = np.sort(np.random.randint(60,100, 25))
```

```
#Generemos fechas aleatoreas.
start_date = pd.to_datetime('2018-06-30')
end_date = pd.to_datetime('2020-11-30')

def gen_rand_date(group, start_date = start_date, end_date = end_date):
    # step 1 in description above:
    days = (end_date - start_date).days
    # step 2:
    d = pd.np.sort(pd.np.random.choice(range(days), len(group)))
    # step 3:
    return start_date + pd.to_timedelta(d,unit='D')

df['fecha'] = df.transform(lambda x: gen_rand_date(x))
```

```
#Generemos materias aleatoreas.
materias = df.materia.apply(lambda x: np.random.choice(['org de datos', 'algoritmos','estadistica','cocina Africana intensiva']) )
df.materia = materias
```

```
#Generemos padrones aleatoreos.
padrones = df.padron.apply(lambda x: np.random.choice(['1010', '1020','1030']) )
df.padron = padrones
```

```
# A. estadisticas de las notas incluyendo el promedio
df.describe() 
```

```
 # A.1 grafico de distribucion de las notas
plt.hist(df.nota,bins=10)
```

```
# B. Cual es la nota mas alta y baja del 2019?
df[df['fecha'].dt.year == 2019].describe()
```

```
# C. Cu√°l es el padr√≥n con mayor cantidad de materias aprobadas (>75) durante el √∫ltimo cuatrimestre
df[df.nota > 75].groupby('padron')['nota'].count().sort_values(ascending=False)
```

```
# D. Cu√°l es la nota promedio
df.nota.mean()
```

```
# E. Cu√°l es la nota promedio por padr√≥n.
df.groupby('padron')['nota'].mean()
```
